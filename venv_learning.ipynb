# Learning Python Virtual Environments (venv)

This notebook covers the essentials of Python virtual environments using the built-in `venv` module. Virtual environments are crucial for managing project dependencies, avoiding conflicts, and ensuring reproducibility in Python development.

## What is a Virtual Environment?

- A virtual environment is an isolated Python environment that allows you to install packages without affecting the global Python installation.
- It helps manage dependencies per project, preventing version conflicts.
- `venv` is Python's built-in module for creating virtual environments (Python 3.3+).

## Activating a Virtual Environment

Activation modifies your shell's PATH to use the virtual environment's Python and pip.

**On macOS/Linux**: `source myenv/bin/activate`

**On Windows**: `myenv\Scripts\activate`

You'll see `(myenv)` in your prompt when activated.

## Installing Packages

Once activated, use `pip` to install packages. They install into the virtual environment.

**Commands**:
- `pip install package_name`
- `pip install -r requirements.txt` (for multiple packages)

```python
# Example: Install requests (run in terminal after activating)
# pip install requests
# Then check installed packages
import subprocess
result = subprocess.run(['pip3', 'list'], capture_output=True, text=True)
print(result.stdout)

```

## Managing Requirements

Freeze your environment's packages to a file for reproducibility.

**Commands**:
- `pip freeze > requirements.txt` (save packages)
- `pip install -r requirements.txt` (install from file)

```python
# Create requirements.txt (run in terminal: pip freeze > requirements.txt)
# Then read it
if os.path.exists('requirements.txt'):
    with open('requirements.txt', 'r') as f:
        print(f.read())
else:
    print('Create requirements.txt first')
    #subprocess.run(['pip3', 'freeze', '>', 'requirements.txt'], shell=True)  # Create requirements.txt using pip freeze
    subprocess.run('pip3 freeze > requirements.txt', shell=True)
# why subprocess.run(['pip3', 'freeze', '>', 'requirements1.txt'], shell=True) is not creating the requirements.txt file? The command subprocess.run(['pip3', 'freeze', '>', 'requirements1.txt'], shell=True) is not creating the requirements.txt file because the redirection operator (>) is a feature of the shell, and when you use subprocess.run with shell=True, it does not interpret the redirection operator as expected. Instead, it treats it as part of the command string, which is why the file is not being created. To create the requirements.txt file using pip freeze, you can use the following code: 
# with open('requirements.txt', 'w') as f:
#     subprocess.run(['pip3', 'freeze'], stdout=f)  # Redirect output to requirements.txt       
# This code opens the requirements.txt file in write mode and then runs the pip freeze command, redirecting its output to the file. This way, the requirements.txt file will be created with the list of installed packages.
# Alternatively, you can run the command in using subprocess.run with string and shell=True like this:      
# subprocess.run('pip3 freeze > requirements.txt', shell=True)  
# This will execute the command in the shell, allowing the redirection operator to work correctly and create the requirements.txt file with the output of pip freeze.   
  
```

## Deactivating a Virtual Environment

To exit the virtual environment and return to the global Python.

**Command**: `deactivate`

Your prompt will return to normal.

```python
# Check deactivation (run after deactivating)
print('Python executable after deactivate:', sys.executable)
```

## Removing a Virtual Environment

Simply delete the environment directory.

**Command**: `rm -rf myenv` (Linux/macOS) or `rmdir /s myenv` (Windows)

```python
# Check if removed
if not os.path.exists('myenv'):
    print('Virtual environment removed')
else:
    print('Still exists')
```

## Best Practices

- Always use virtual environments for projects.
- Don't commit the `myenv` directory to version control; use `requirements.txt`.
- Use descriptive names for environments.
- Regularly update packages with `pip install --upgrade package`.
