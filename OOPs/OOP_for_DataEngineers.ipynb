## Section 1: OOP Concepts for Data Engineers

### Key Definitions

**Class**: A blueprint for creating objects (think: data schema)
- Define structure and behavior
- Example: `DataRecord`, `CSVReader`, `DataValidator`

**Object**: An instance of a class (think: actual data record)

- Example: A single customer record from a dataset

**Attributes**: Variables that hold data (think: columns in a table)

**Methods**: Functions that operate on data (think: transformations)

## Section 3: Encapsulation with Data Models

**Encapsulation** = Protecting data by hiding it and controlling access through methods

### Why It Matters for Data Engineers
- Validate data before storing
- Prevent invalid states (negative amounts, empty required fields)
- Control how external code accesses internal data
- Makes refactoring easier without breaking other code

### Implementation: Private Attributes & Getter/Setter Methods

````python
# Encapsulation Example: Protected Data with Validation
class Transaction:
    """A transaction class with data validation"""
    
    def __init__(self, transaction_id, amount, status="pending"):
        self._id = transaction_id  # Private attribute (convention: prefix _)
        self._amount = None
        self._status = status
        self.set_amount(amount)  # Use setter for validation
    
    def set_amount(self, amount):
        """Setter: Validate before storing"""
        if amount < 0:
            raise ValueError("Amount cannot be negative!")
        if amount > 1_000_000:
            raise ValueError("Amount exceeds maximum limit!")
        self._amount = amount
    
    def get_amount(self):
        """Getter: Provide controlled access"""
        return self._amount
    
    def get_id(self):
        return self._id
    
    def update_status(self, new_status):
        """Update status with validation"""
        valid_statuses = ["pending", "success", "failed"]
        if new_status not in valid_statuses:
            raise ValueError(f"Invalid status. Must be one of {valid_statuses}")
        self._status = new_status
    
    def __str__(self):
        """String representation of the object"""
        return f"Transaction({self._id}: ${self._amount:.2f}, {self._status})"

# Usage
try:
    txn = Transaction("TXN001", 500, "pending")
    print(txn)
    ########## Discussion on __str__ method
    # What does __str__ do? It provides a human-readable string representation of the object.
    # can you explain how it works? When we print the object, Python calls the __str__ method to get a string representation of the object. This allows us to see a more informative output instead of just the memory address of the object.
    # do we need to define __str__ for all classes? No, it's not mandatory, but it is a good practice to define it for classes where a human-readable representation is useful for debugging or logging purposes.
    # because in one previous cell where we defined datarecord class, we did not define __str__, so when we printed the object, it showed something like <__main__.DataRecord object at 0x7f8b8c8d0>, which is not very informative. By defining __str__, we can control how the object is represented as a string, making it easier to understand the output when we print the object.

    ######### diff between self._amount and self.set_amount(amount) in __init__?
    # In __str__, we return a formatted string that includes the self._amount which is set to None in __init__. then when we print the object why it returned the correct amount? 
    # When you instantiate the class with Transaction("TXN001", 500), Python executes the __init__ method as a linear sequence of instructions. It first assigns the ID and initializes self._amount to None as a temporary placeholder in the object's memory. However, the process doesn't stop there; 
    # the next line explicitly calls the self.set_amount(amount) method, passing the value 500 into it. At this point, the program's execution "jumps" from the constructor to the setter method, where it performs the validation checks. Since 500 passes those checks, the setter executes the assignment self._amount = amount,
    #  which overwrites the initial None with the actual numerical value. Because this entire sequence happens before the object is fully finished being created, by the time you call print(txn), the internal self._amount has already been updated to 500, allowing the __str__ method to access and display the correct validated data.


    
    # Using getter
    print(f"Amount: ${txn.get_amount()}")

    txn._amount = -1000  # Direct access (not recommended)
    print(f"Directly updated amount (not recommended): ${txn.get_amount()}")
    # This is a problem because it bypasses the validation logic in the setter, allowing us to set an invalid amount. This is why we should use the setter method to ensure data integrity. 
    # it is understood that if a variable starts with an underscore, it is a convention to indicate that it is intended for internal use and should not be accessed directly from outside the class. However, this is just a convention and does not enforce any actual access restrictions. Therefore, while you can technically access and modify _amount directly, it is not recommended because it can lead to unintended consequences and data integrity issues, as it bypasses the validation logic defined in the setter method.

    
    # Using setter
    txn.set_amount(750)
    print(f"Updated: {txn}")
    
    # This will fail - encapsulation protects data
    #txn.set_amount(-100)  # ValueError!
    
    print("\n✓ Encapsulation ensures data integrity!")
    
except ValueError as e:
    print(f"Error: {e}")
````

## Section 4: Inheritance for Reusable Pipelines

**Inheritance** = Create specialized classes from a base class (parent → child)

### Why It Matters for Data Engineers
- Define a base `DataReader` class, then create `CSVReader`, `JSONReader`, `ParquetReader`
- Share common logic (connect, validate schema, handle errors)
- Reduce code duplication
- Easy to add new file format readers

### Example: ETL Pipeline Components

````python
# Inheritance Example: Data Reader Pipeline
from abc import ABC, abstractmethod

class DataReader(ABC):
    """Base class for all data readers"""
    
    def __init__(self, source_path):
        self.source_path = source_path
        self.data = None

    
    @abstractmethod
    def read(self):
        """Must be implemented by child classes"""
        pass
    file_extension = None  # Class variable to be defined in child classes


    def validate_path(self):
        """Common method for all readers"""
        
        if not self.source_path.endswith(self.file_extension):
            raise ValueError(f"Invalid file format. Expected a {self.file_extension}")

class CSVReader(DataReader):
    """Child class for reading CSV files"""
    file_extension = ".csv"
    def read(self):
        
        
        self.validate_path()
        # Simulating CSV read
        print(f"Reading CSV from {self.source_path}")
        self.data = [
            {"name": "Alice", "amount": 1500},
            {"name": "Bob", "amount": 2000}
        ]
        return self.data

class JSONReader(DataReader):
    """Child class for reading JSON files"""
    file_extension = ".json"
    def read(self):
        self.validate_path()
        # Simulating JSON read
        print(f"Reading JSON from {self.source_path}")
        self.data = [
            {"name": "Charlie", "amount": 3000},
            {"name": "Diana", "amount": 2500}
        ]
        return self.data

class ParquetReader(DataReader):
    """Child class for reading Parquet files"""
    file_extension = ".parquet"
    def read(self):
        self.validate_path()
        # Simulating Parquet read
        print(f"Reading Parquet from {self.source_path}")
        self.data = [
            {"name": "Eve", "amount": 4000},
            {"name": "Frank", "amount": 3500}
        ]
        return self.data

# Usage: Different readers, same interface
readers = [
    CSVReader("data.csv"),
    JSONReader("data.json"),
    ParquetReader("data.parquet")
]

print("🔄 Processing data from different sources:\n")
for reader in readers:
    data = reader.read()
    print(f"  Loaded {len(data)} records\n")

print("✓ Inheritance allows code reuse and consistent interfaces!")


````

## Section 5: Polymorphism in Data Processing

**Polymorphism** = "Many forms" - same method, different implementations based on object type

### Why It Matters for Data Engineers
- Write generic processing functions that work with any data processor
- Add new data sources without modifying existing code (Open/Closed Principle)
- Process CSV, JSON, Parquet identically through a shared interface

### Example: Processing Different Data Sources

````python
# Polymorphism Example: Data Processors with Shared Interface
class DataProcessor(ABC):
    """Base class for all data processors"""
    
    @abstractmethod
    def process(self, data):
        pass
    
    @abstractmethod
    def get_format_name(self):
        pass

class CSVProcessor(DataProcessor):
    def process(self, data):
        return f"Processing {len(data)} CSV records: {data}"
    
    def get_format_name(self):
        return "CSV"

class JSONProcessor(DataProcessor):
    def process(self, data):
        return f"Processing {len(data)} JSON records: {data}"
    
    def get_format_name(self):
        return "JSON"

class DatabaseProcessor(DataProcessor):
    def process(self, data):
        return f"Writing {len(data)} records to database: {data}"
    
    def get_format_name(self):
        return "Database"

# Generic function that works with ANY processor (Polymorphism!)
def run_etl_pipeline(processor: DataProcessor, data):
    """
    Generic ETL function that accepts any processor
    Notice: We don't care about the specific processor type!
    """
    print(f"Starting ETL for {processor.get_format_name()}...")
    result = processor.process(data)
    print(result)
    print("✓ Pipeline complete\n")

# Usage: Same function, different processors
sample_data = [
    {"id": 1, "name": "Alice"},
    {"id": 2, "name": "Bob"}
]

processors = [
    CSVProcessor(),
    JSONProcessor(),
    DatabaseProcessor()
]

print("🔄 Polymorphism in action - same function, different behaviors:\n")
for processor in processors:
    run_etl_pipeline(processor, sample_data)

print("✓ Polymorphism allows flexible, extensible code!")

'''
1. Polymorphism: "Many Shapes, One Interface"
Inheritance is about sharing code. Polymorphism is about swappability.
In your DataReader example, you had a list: [CSVReader(), JSONReader()]. When you looped through them and called .read(), you didn't care which class it was. That is Polymorphism.
Inheritance: "I am a DataReader, so I get the validate_path tool for free."
Polymorphism: "I don't care if you are a CSV or JSON reader; as long as you have a .read() button, I can use you."
Data Engineering Example: You write a pipeline that accepts a "Database Connector." It doesn't matter if the object passed is PostgresConnector or SnowflakeConnector. The pipeline just calls .connect().
'''
````

## Section 6: Composition and Dependency Injection

**Composition** = Building complex objects from simpler objects (HAS-A relationship)

vs. **Inheritance** = IS-A relationship (child is a type of parent)

### Why It Matters for Data Engineers
- Build flexible pipelines by composing validators, readers, transformers
- Inject dependencies (like validators) instead of hardcoding
- Easier to test and modify behavior

### Example: Composing Data Pipeline Components

````python
# Composition & Dependency Injection Example
class Validator(ABC):
    @abstractmethod
    def validate(self, data):
        pass

class NotNullValidator(Validator):
    def validate(self, data):
        if not data:
            raise ValueError("Data cannot be null")
        return True

class RangeValidator(Validator):
    def __init__(self, min_val, max_val):
        self.min_val = min_val
        self.max_val = max_val
    
    def validate(self, value):
        if not (self.min_val <= value <= self.max_val):
            raise ValueError(f"Value {value} not in range [{self.min_val}, {self.max_val}]")
        return True

# Pipeline that COMPOSES validators (has-a relationship)
class DataValidationPipeline:
    def __init__(self, validators: list):
        """Dependency Injection: inject validators as a list"""
        self.validators = validators
    
    def validates(self, data):
        """Run all validators"""
        for validator in self.validators:
            validator.validate(data)
        return data
    
    def add_validator(self, validator):
        """Add a validator dynamically"""
        self.validators.append(validator)

# Usage: Build flexible pipelines
print("Composition & Dependency Injection Example:\n")

# Create validators
null_check = NotNullValidator()
range_check = RangeValidator(0, 100)

# Inject validators into pipeline
pipeline = DataValidationPipeline([null_check, range_check])

# Test valid data
try:
    result = pipeline.validates(50)
    print(f"✓ Data {result} passed validation")
except ValueError as e:
    print(f"✗ Validation failed: {e}")

# Test invalid data
try:
    result = pipeline.validates(150)
    print(f"✓ Data {result} passed validation")
except ValueError as e:
    print(f"✗ Validation failed: {e}")

# Add a new validator dynamically
class LengthValidator(Validator):
    def __init__(self, max_length):
        self.max_length = max_length
    
    def validate(self, data):
        if len(str(data)) > self.max_length:
            raise ValueError(f"Data length exceeds {self.max_length}")
        return True

length_check = LengthValidator(2)
pipeline.add_validator(length_check)
print("\n✓ Composition allows flexible, reusable pipelines!")

'''
Composition: "Has-a" vs "Is-a"
This is the biggest rival to Inheritance.
Inheritance (Is-a): A CSVReader is a DataReader. (Rigid relationship).
Composition (Has-a): A DataPipeline has a Reader and has a Writer.
'''
````

### Practice 1: Build a Data Record Class with Validation

**Problem**: Create a `CustomerRecord` class for a customer database that:
- Stores: id, name, email, signup_date, lifetime_value
- Validates: email format, non-negative lifetime_value
- Provides: method to convert to dictionary, string representation
- Ensures: data integrity through encapsulation

**Requirements**:
✓ Private attributes with getters/setters
✓ Email validation (contains @)
✓ Positive amount validation
✓ `to_dict()` method for database insertion
✓ `__str__()` method for logging

**Your Solution**:

````python
# PRACTICE 1 SOLUTION: CustomerRecord Class
from datetime import datetime

class CustomerRecord:
    """A customer record with validation and data integrity"""
    
    def __init__(self, customer_id, name, email, signup_date, lifetime_value):
        self._id = customer_id
        self._name = name
        self._email = None
        self._signup_date = signup_date
        self._lifetime_value = None
        
        # Use setters for validation
        self.set_email(email)
        self.set_lifetime_value(lifetime_value)
    
    def set_email(self, email):
        if "@" not in email or "." not in email.split("@")[1]:
            raise ValueError(f"Invalid email format: {email}")
        self._email = email
    
    def set_lifetime_value(self, value):
        if value < 0:
            raise ValueError("Lifetime value cannot be negative")
        self._lifetime_value = value
    
    def get_id(self):
        return self._id
    
    def get_name(self):
        return self._name
    
    def get_email(self):
        return self._email
    
    def get_lifetime_value(self):
        return self._lifetime_value
    
    def to_dict(self):
        """Convert record to dictionary for database insertion"""
        return {
            "customer_id": self._id,
            "name": self._name,
            "email": self._email,
            "signup_date": self._signup_date,
            "lifetime_value": self._lifetime_value
        }
    
    def __str__(self):
        return f"Customer({self._id}: {self._name}, {self._email}, ${self._lifetime_value:.2f})"

# Test the solution
print("Testing CustomerRecord Class:\n")

try:
    customer = CustomerRecord("C001", "Alice Smith", "alice@dataeng.com", "2023-01-15", 2500.00)
    print(f"✓ Created: {customer}")
    print(f"✓ As Dict: {customer.to_dict()}")
    
    # Try to set invalid email
    customer.set_email("invalid-email")
except ValueError as e:
    print(f"✗ Validation caught error: {e}")

print("\n✓ Practice 1 Complete!")
````

### Practice 2: Build File Processors Using Inheritance

**Problem**: Create a base `FileProcessor` class and subclasses for different formats:
- `CSVProcessor`: reads and counts records
- `JSONProcessor`: reads and extracts keys
- `ParquetProcessor`: reads and shows data types

**Requirements**:
✓ Abstract base class with common methods
✓ Each subclass implements specific `read()` logic
✓ Shared `get_record_count()` method
✓ Each has format-specific processing methods

**Your Solution**:

````python
# PRACTICE 2 SOLUTION: File Processors with Inheritance
class FileProcessor(ABC):
    """Base class for file processors"""
    
    def __init__(self, filepath):
        self.filepath = filepath
        self.data = None
    
    @abstractmethod
    def read(self):
        """Must be implemented by subclasses"""
        pass
    
    def get_record_count(self):
        """Shared method - works for all subclasses"""
        if self.data is None:
            return 0
        return len(self.data)

class CSVProcessor(FileProcessor):
    def read(self):
        # Simulating CSV read
        self.data = [
            {"name": "Alice", "amount": 1500},
            {"name": "Bob", "amount": 2000},
            {"name": "Charlie", "amount": 1800}
        ]
        return self.data
    
    def get_columns(self):
        if not self.data:
            return []
        return list(self.data[0].keys())

class JSONProcessor(FileProcessor):
    def read(self):
        # Simulating JSON read
        self.data = [
            {"id": 1, "user": "Diana", "score": 95},
            {"id": 2, "user": "Eve", "score": 87}
        ]
        return self.data
    
    def get_schema(self):
        if not self.data:
            return {}
        first_record = self.data[0]
        return {key: type(value).__name__ for key, value in first_record.items()}

class ParquetProcessor(FileProcessor):
    def read(self):
        # Simulating Parquet read
        self.data = [
            {"timestamp": "2024-01-01", "value": 100, "flag": True},
            {"timestamp": "2024-01-02", "value": 200, "flag": False}
        ]
        return self.data
    
    def get_data_types(self):
        if not self.data:
            return {}
        first = self.data[0]
        return {k: type(v).__name__ for k, v in first.items()}

# Test the processors
print("Practice 2: File Processors with Inheritance\n")

processors = [
    CSVProcessor("data.csv"),
    JSONProcessor("data.json"),
    ParquetProcessor("data.parquet")
]

for processor in processors:
    processor.read()
    print(f"{processor.__class__.__name__}:")
    print(f"  Records: {processor.get_record_count()}")
    
    if hasattr(processor, 'get_columns'):
        print(f"  Columns: {processor.get_columns()}")
    
    if hasattr(processor, 'get_schema'):
        print(f"  Schema: {processor.get_schema()}")
    
    if isinstance(processor, ParquetProcessor):
        print(f"  Types: {processor.get_data_types()}")
    print()

print("✓ Practice 2 Complete!")
````

### Practice 3: Use Polymorphism in ETL Pipeline

**Problem**: Create an ETL pipeline that accepts any data transformer:
- Base `DataTransformer` class
- `UpperCaseTransformer`: converts names to uppercase
- `AggregationTransformer`: sums amounts by key
- `FilterTransformer`: filters records by condition

Create a generic `run_transformation()` function that works with any transformer.

**Requirements**:
✓ Abstract base class `DataTransformer`
✓ At least 3 concrete transformers
✓ Generic function accepts any transformer
✓ Process same data with different transformers

**Your Solution**:

````python
# PRACTICE 3 SOLUTION: Polymorphic Data Transformers
class DataTransformer(ABC):
    """Abstract base class for data transformers"""
    
    @abstractmethod
    def transform(self, data):
        """Must be implemented by subclasses"""
        pass

class UpperCaseTransformer(DataTransformer):
    """Convert names to uppercase"""
    
    def transform(self, data):
        result = []
        for record in data:
            record['name'] = record['name'].upper()
            result.append(record)
        return result

class AggregationTransformer(DataTransformer):
    """Sum amounts by name"""
    
    def transform(self, data):
        aggregated = {}
        for record in data:
            name = record['name']
            amount = record['amount']
            aggregated[name] = aggregated.get(name, 0) + amount
        
        return [{'name': k, 'total': v} for k, v in aggregated.items()]

class FilterTransformer(DataTransformer):
    """Filter records with amount > threshold"""
    
    def __init__(self, threshold=1500):
        self.threshold = threshold
    
    def transform(self, data):
        return [r for r in data if r['amount'] >= self.threshold]

# Generic ETL function - works with ANY transformer
def run_transformation(data, transformer: DataTransformer):
    """
    Generic transformation function
    Demonstrates polymorphism - accepts any DataTransformer subclass
    """
    print(f"Running {transformer.__class__.__name__}...")
    result = transformer.transform(data)
    print(f"Result: {result}\n")
    return result

# Test data
sample_data = [
    {'name': 'alice', 'amount': 1500},
    {'name': 'bob', 'amount': 2000},
    {'name': 'alice', 'amount': 1000},
    {'name': 'charlie', 'amount': 1800}
]

print("Practice 3: Polymorphism in ETL Pipeline\n")
print(f"Original Data: {sample_data}\n")

# Apply different transformers using the same function
run_transformation(sample_data, UpperCaseTransformer())
run_transformation(sample_data, AggregationTransformer())
run_transformation(sample_data, FilterTransformer(threshold=1800))

print("✓ Practice 3 Complete!")
````

| Pattern | Use Case | Example |

| **Base Class + Inheritance** | Readers for different formats | `DataReader` → `CSVReader`, `JSONReader` |

- **Maintainability**: Would someone understand this code 6 months from now?
- **Composition**: Can I build complex pipelines from simple components?

## Bonus: Common OOP Mistakes & How to Avoid Them

### ❌ Mistake 1: Over-Engineering with Too Many Classes
**Problem**: Creating a class for everything (even simple functions)

**Solution**: Ask "Will this be reused or need extensions?" If no, keep it simple.

### ❌ Mistake 2: Not Using Inheritance When Appropriate
**Problem**: Copying the same code across multiple reader classes

**Solution**: Extract common logic into a base class that specific readers inherit from.

### ❌ Mistake 3: Breaking Encapsulation
**Problem**: Directly accessing `reader.raw_data` instead of using proper methods

**Solution**: Always use getters/setters; it gives you control over validation and change.

### ❌ Mistake 4: Tight Coupling (Classes Depending Too Much)
**Problem**: `Pipeline` hardcoded to use `CSVReader` only

**Solution**: Accept any `DataReader` subclass; use composition and dependency injection.

### ❌ Mistake 5: Ignoring the Single Responsibility Principle
**Problem**: A `DataValidator` class that also transforms and writes data

**Solution**: Each class should do ONE thing well. Use composition to combine.
